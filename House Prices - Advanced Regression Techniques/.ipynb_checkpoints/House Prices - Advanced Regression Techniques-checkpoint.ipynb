```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.utils import resample
```

```python
train = pd.read_csv("Sourses/train.csv")
test = pd.read_csv("Sourses/test.csv")
train.head()
```

```python
train.columns
```

```python
features = ["LotArea", "MSSubClass", "PoolArea", "YearBuilt", "MoSold", "YrSold"]
X_train = train[features]
X_test = test[features]
y_train = train["SalePrice"]
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
test
```

```python
submission = test[["Id"]]
submission
```

```python
submission["SalePrice"] = predictions
```

```python
params = pd.Series(model.coef_, index=X_train.columns)
params
```

```python
np.random.seed(1)
err = np.std([model.fit(*resample(X_train, y_train)).coef_
              for i in range(1000)], 0)
```

```python
print(pd.DataFrame({'effect': params.round(0),
                    'error': err.round(0)}))
```

```python
submission
```

```python
submission.to_csv("submission.csv", index = None)
```
